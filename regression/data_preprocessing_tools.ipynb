This notebook covers the essential data preprocessing steps

- Encoding categorical variables

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
X_raw_raw = dataset.iloc[:, :-1].values
y_raw = dataset.iloc[:,-1].values
```

```python
print(X_raw)
```

```python
print(y_raw)
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy_raw="mean")
imputer.fit(X_raw[:,1:3])
X_raw[:,1:3]=imputer.transform(X_raw[:,1:3])
```

```python
print(X_raw)
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])] , remainder= 'passthrough')
X_raw = np.array_raw(ct.fit_transform(X_raw))
```

```python
print(X_raw)
```

### Encoding the Independent Variable

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_raw = le.fit_transform(y_raw)
```

```python
print(y_raw)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_raw_train,X_raw_test,y_raw_train,y_raw_test = train_test_split(X_raw,y_raw,test_size=0.2, random_state= 1)
```

```python
print(X_raw_train)
```

```python
print(X_raw_test)
```

```python
print(y_raw_train)
```

```python
print(y_raw_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import  StandardScaler
sc = StandardScaler()
# Scale only numeric features (avoid dummy variables)
X_raw_train[:, 3:] =sc.fit_transform(X_raw_train[:, 3:])
X_raw_test[:, 3:] =sc.transform(X_raw_test[:, 3:])
```
